# libraries

```python
from skimage import color, data, filters , measure
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def plot_comparison(original, filtered, title_filtered,title_original='original'):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 10), sharex=True,
    sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title(title_original)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered)
    ax2.axis('off')
```

```python
def plot_contrast(org_image,eq_image):
    fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(8, 8))
    ax[0,0].imshow(org_image,cmap=plt.cm.gray)
    ax[0,0].set_title('original')
    ax[0,0].axis('off')

    ax[0,1].hist(org_image.ravel(), bins=256)
    ax[0,1].set_title('Histogram of image')


    # Show the resulting image
    ax[1,0].imshow(eq_image,cmap=plt.cm.gray)
    ax[1,0].set_title('after costrast by equalization')
    ax[1,0].axis('off')

    ax[1,1].hist(eq_image.ravel(), bins=256)
    ax[1,1].set_title('Equalization Histogram')
```

```python
def show_image_contour(image, contours):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 8))
    ax1.imshow(image, cmap='gray_r')
    ax1.set_title('original')
    ax1.axis('off')
    
    for n, contour in enumerate(contours):
        ax2.plot(contour[:, 1], contour[:, 0], linewidth=3)
    ax2.imshow(image, interpolation='nearest', cmap='gray_r')
    ax2.set_title('Contours')
    ax2.axis('off')
```

```python
def show_image_with_corners(image, coords, title="Corners detected"):
    plt.imshow(image, interpolation='nearest', cmap='gray')
    plt.title(title)
    plt.plot(coords[:, 1], coords[:, 0], '+r')
    plt.axis('off')
    plt.show()
```

# Image Restoration

- **uses**
    - Fixing damaged images
    - Text removing
    - Logo removing
    - Object removing
    
    
- **inpainting** Reconstructing lost or deteriorated parts of images.
- It needs the location of the damaged pixels to be filled, as a mask image on top of the image to work with.
- A **mask image** is simply an image where some of the pixel intensity values are zero, and others are non-zero. 

## Fixing damaged images

```python
def get_mask(image):
#Creates mask with three defect regions 
    mask = np.zeros(image.shape[:-1])
    mask[101:106, 0:240] = 1
    mask[152:154, 0:60] = 1
    mask[153:155, 60:100] = 1
    mask[154:156, 100:120] = 1
    mask[155:156, 120:140] = 1
    mask[212:217, 0:150] = 1
    mask[217:222, 150:256] = 1
    return mask
```

```python
defect_image=plt.imread('./images/chapter3/damaged_astronaut.png')

# Import the module from restoration
from skimage.restoration import inpaint

# Show the defective image
show_image(defect_image, 'Image to restore')

# Apply the restoration function to the image using the mask
mask=get_mask(defect_image)
restored_image = inpaint.inpaint_biharmonic(defect_image,
                                            mask,
                                            multichannel=True)
show_image(restored_image)
```

## Removing logos

```python
image_with_logo=plt.imread('./images/chapter3/with-new-logo.png')

# Initialize the mask
mask = np.zeros(image_with_logo.shape[:-1])

# Set the pixels where the logo is to 1
mask[210:290, 360:425] = 1

# Apply inpainting to remove the logo
image_logo_removed = inpaint.inpaint_biharmonic(image_with_logo,
                                  mask,
                                  multichannel=True)

# Show the original and logo removed images
plot_comparison(image_with_logo,image_logo_removed, 'Image with logo removed')
```

# Segmentation

- The goal is to partition images into regions, or segments, to simplify and/or change the representation into something more meaningful and easier to analyze.
- For example, before a tumor is analyzed in a computed tomography, it has to be detected and somehow isolated from the rest of the image.
- before recognizing a face, it has to also be picked out from its background
- Previously we learned about Thresholding, which is the simplest method of segmentation. Separating foreground from background.

# Superpixels
- A single pixel, standing alone by itself, is not a natural representation.
- So, we can explore more logical meanings in an image that's formed by bigger regions or grouped pixels.These are known as superpixels.
- A superpixel is a group of connected pixels with similar colors or gray levels. These carry more meaning than their simple pixel grid counterparts.

**Benefits of superpixels**
- used in visual tracking and image classification 
- reduce an image from thousands of pixels down to some regions for subsequent algorithms, so you can compute features on more meaningful regions and have computational efficiency.


**Two types of segmentation**
1. Supervised (thresholding)
    - where some prior knowledge is used to guide the algorithm. Like the kind of thresholding in which we specify the threshold value ourselves.
2. unsupervised (based on superpixels)
    - where no prior knowledge is required. These algorithms try to subdivide images into meaningful regions automatically. The user may still be able to tweak certain settings to obtain the desired output. Like the **otsu thresholding** .
    
    
## Simple Linear Iterative Clustering (SLIC)
- Simple Linear Iterative Clustering or SLICis  a unsupervised segmentation technique based on superpixels.
- It segments the image using a machine learning algorithm called **K-Means clustering**.
- It takes in all the pixel values of the image and tries to separate them into a predefined number of sub-regions.

```python
face_image=plt.imread('./images/chapter3/chinese.jpg')

# Import the slic function from segmentation module
from skimage.segmentation import slic

# Import the label2rgb function from color module
from skimage.color import label2rgb

# Obtain the segmentation with 400 regions
segments = slic(face_image, n_segments= 400)

# Put segments on top of original image to compare
segmented_image = label2rgb(segments, face_image, kind='avg')

# Show the segmented image
plot_comparison(face_image,segmented_image.astype('uint8'),"Segmented image, 400 superpixels")
```

```python
coffee=data.coffee()

segments = slic(coffee, n_segments=400)

segmented_image = label2rgb(segments, coffee, kind='avg')

plot_comparison(coffee,segmented_image.astype('uint8'),"Segmented image, 400 superpixels")
```

# contours
- A contour is a closed shape of points or line segments, representing the boundaries of these objects.
- Once we find the contours, we can identify the total points in domino tokens, exactly what we do in this example, where we count 29!
- So we can measure size, classify shapes or determining the number of objects in an image.

- **Steps for finding contour**
    1. transform to **grayscale**, If it's colored.
    2. Obtain the optimal **thresh value**.
    3. Apply thresholding and obtain the **binary image**.
    4. call the **find_contours()** function and set a constant level value.
       
       
- **find contours()**
    - This function finds the contour lines or joins points(pixels) of equal elevation or brightness in a 2D array above a given level value.
    - parameters: thresholded image ,  constant level value.
    - The function returns a list with all contours of the image. With the coordinates along the contour.
    
    
- **Constant level value**
  - The level value varies between 0 and 1
  - the closer to 1 the more sensitive the method is to detecting contours, so more complex contours will be detected. 


- **contour's shape**
    - After executing these steps we obtain a list of contours. Each contour is an ndarray of shape (n, 2), consisting of n row and column coordinates along the contour.     

```python
image_dice=plt.imread('./images/chapter3/dices.png')

# 1- Make the image grayscale
image_dice = color.rgb2gray(image_dice)

# 2- Obtain the optimal thresh value
thresh = filters.threshold_otsu(image_dice)

# 3- Apply thresholding
binary = image_dice > thresh

# 4- Find contours at a constant value of 0.8
contours =measure.find_contours(binary, 0.8)

# Show the image
show_image_contour(image_dice, contours)
print('No. of contours = ',len(contours))
```
