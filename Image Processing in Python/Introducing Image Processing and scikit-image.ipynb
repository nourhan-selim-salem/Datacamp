# libraries

```python
from skimage import color, data, filters , measure
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import numpy as np
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def plot_comparison(original, filtered, title_filtered,title_original='original'):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 10), sharex=True,
    sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title(title_original)
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(title_filtered)
    ax2.axis('off')
```

# scikit-image

```python
# Load the rocket image
rocket = data.rocket()
```

## Gray images

```python
# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)

plot_comparison(rocket,gray_scaled_rocket, 'Grayscale image')
```

## Flipping

## Color histograms
- The histogram of an image is a graphical representation of the amount of pixels of each intensity value, From 0 (pure black) to 255(pure white). 
- We can also create histograms from RGB-3 colored images. 
- Histograms are used:
    - to threshold images 
    - to alter brightness and contrast 
    - to equalize an image 

```python
# Obtain the red channel
red_channel = image[:, :, 0]

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# Set title and show
plt.title('Red Histogram')
plt.show()
```

# Thresholding
- 255 (white) if pixel > thresh value
- 0 (black) if pixel < thresh value
- Simplest method of image segmentation (Object detection)

```python
# Obtain the optimal threshold value
image = color.rgb2gray(data.camera())
thresh = 127
# Apply thresholding to the image
binary = image > thresh
# Show the original and thresholded
plot_comparison(image ,binary, 'Thresholded')
```

## Inverted thresholding

```python
# Apply thresholding to the image
inverted_binary = image <= thresh
# Show the original and thresholded
plot_comparison(image ,inverted_binary,'Inverted thresholded')
```

## Optimal thresh value
### Global

```python
# Import the otsu threshold function
from skimage.filters import threshold_otsu
# Obtain the optimal threshold value
thresh = threshold_otsu(image)
# Apply thresholding to the image
binary_global = image > thresh

plot_comparison(image ,binary_global, 'Local thresholding')
```
